### $O(1)$  -  (constant)

```python
def square(x):
    return x * x

square(3)
```

```python
def first(x): # x is a list
    start = time.time()
    r = x[0]
    t = time.time() - start
    return r, len(x), t

def middle(x):
    start = time.time()
    r = x[len(x)//2]
    t = time.time() - start
    return r, len(x), t

def last(x):
    start = time.time()
    r = x[-1]
    t = time.time() - start
    return r, len(x), t
```

```python
random_lists = random_list(10000)
rst = [last(l) for l in random_lists]
len(rst)
rst
```

```python
x = list(zip(*rst))[1]
y = list(zip(*rst))[2]

plt.plot(x, y)
```

### $O(lgn)$

```python
import bisect
def bs(nums, target):
    sorted(nums)
    start = time.time()
    i = bisect.bisect_left(nums, target)
    if i != len(nums) and nums[i] == target:
        t = time.time() - start
        return i, len(nums), t
    t = time.time() - start
    return -1, len(nums), t
```

```python
random_lists = random_list(10000)
```

```python
rst = [bs(l, 100) for l in random_lists]
len(rst)
rst
```

```python
x = list(zip(*rst))[1]
y = list(zip(*rst))[2]

plt.plot(x, y)
```

### $O(n)$

```python
def find_max(l):
    start = time.time()
    
    if l == None:
        return None
    mx = l[0]
    for n in l:
        if n > mx:
            mx = n
            
    t = time.time() - start
    
    return mx, len(l), t
```

```python
random_lists = random_list(20000)
rst = [find_max(l) for l in random_lists]
len(rst)
rst
```

```python
x = list(zip(*rst))[1]
y = list(zip(*rst))[2]

plt.plot(x, y)
```

### $O(nlgn)$

```python
def mysort(l):
    start = time.time()
    l.sort()
    t = time.time() - start
    return l[0], len(l), t
```

```python
random_lists = random_list(10000)
rst = [mysort(l) for l in random_lists]
len(rst)
rst
```

```python
x = list(zip(*rst))[1]
y = list(zip(*rst))[2]

plt.plot(x, y)
```

### $O(n^2)$

```python
def has_duplicate(l):
    start = time.time()
    
    rst = False
    for i in range(len(l)):
        for j in range(i + 1, len(l)):
            if l[i] == l[j]:
                rst = True
        
    t = time.time() - start
    
    return rst, len(l), t
```

```python
random_lists = random_list(100)
rst = [has_duplicate(l) for l in random_lists]

x = list(zip(*rst))[1]
y = list(zip(*rst))[2]

plt.plot(x, y)
```
